### Question 1 : 
Make a Pie Chart of the visits to Thrill Ride attractions.

**Note:** For this question, display the pie chart in the notebook and print the data used to create the pie chart as a list of lists (ex: [['Ride 1', 10], ['Ride 2', 100], ...])


````python
# Graded Cell, PartID: gtMqY
# your code here

db.execute("Select attraction, count(*) From checkin Group By attraction")
l = db.fetchall()
freq = pd.DataFrame.from_records(l, columns=['RideID', 'Freq'])

db.execute("SELECT AttractionID, name FROM attraction as a where a.category LIKE '%thrill%';")
a_list =  db.fetchall()
attraction_list = pd.DataFrame.from_records(a_list, columns=['RideID', 'RideName'])

new_df=attraction_list.merge(freq[['RideID', 'Freq']])
new_df = new_df.drop(['RideID'], axis =1)

plt.pie(new_df['Freq'], labels=new_df['RideName'], autopct='%0.2f%%')
plt.axis('equal')
plt.title('Thrill Ride Vists Pie Chart\n')
plt.show()
print(new_df.values.tolist())
````

### Question 2 : 
Make a bar chart of total visits to food stalls.

**Note:** For this question, display the bar chart in the notebook and print the data used to create the bar chart as a list of lists (ex: [['Stall 1', 10], ['Stall 2', 50], ...])


````python
# Graded Cell, PartID: 9Ocyl
# your code here
db.execute("Select attraction, count(*) From checkin Group By attraction")
l = db.fetchall()
freq = pd.DataFrame.from_records(l, columns=['Attraction_ID', 'Freq'])

db.execute("SELECT AttractionID, name FROM attraction as a where a.category LIKE '%food%';")
a_list =  db.fetchall()
attraction_list = pd.DataFrame.from_records(a_list, columns=['Attraction_ID', 'Foodstall'])

new_df=attraction_list.merge(freq[['Attraction_ID', 'Freq']])
new_df = new_df.drop(['Attraction_ID'], axis =1)

new_df.plot(kind='bar', x = 'Foodstall', y = 'Freq', fontsize=11)
plt.title('Food Stall Visits\n')
plt.xlabel('Different Food stalls\n')
plt.ylabel('Number of Visitors\n')
plt.show()

print(new_df.values.tolist())
````

### Question 3 : 
Make a line chart of attendance at Atmosfear every five minutes.

**Note:** For this question, display the line chart in the notebook and print the data used to create the chart as a list of lists (ex: [['Stall 1', 10], ['Stall 2', 50], ...]) or tuple lists (ex: [('Stall 1, 10),('Stall 2', 50), ...]

The first item in tuples is irrelevant, but you can put in some meaningful information. The second item is the count of visits at that moment. For example, your output should look like this (in Python syntax; not relevant to the correct answer): [ (0, 0), (1, 7), (2, 3), …, (190, 4), (191, 5) ].

````python
# Graded Cell, PartID: 0zcEV
# your code here
db.execute("SELECT attractionId FROM attraction where attraction.Name = 'Atmosfear';")
a_ID = db.fetchone()[0]
db.execute("SELECT visitorID, sequence FROM sequences where sequence LIKE '%" + str(a_ID) + "%';")
sequences = db.fetchall()

def check(row):
  l = []
  for x in row['sequence'].split("-"):
    if x==str(a_ID):
      l.append(1)
    else:
      l.append(0)
  return l

seq_df = pd.DataFrame.from_records(sequences, columns=['visitor', 'sequence'])
seq_df['s_list'] = seq_df.apply(check, axis = 1)
attendance = np.sum(seq_df['s_list'].values.tolist(), axis=0)

#x_axis = range(0, len(attendance)*5, 5)
x_axis = range(0, len(attendance))
plt.plot(x_axis, attendance)
plt.ylabel('Attendance')
plt.xlabel('Time(min)')
plt.title('Atmosfear Attendance')
plt.show()
result = [[i, attendance[i]] for i in range(len(x_axis))]
print(result)
````

### Question 4: 
Make a box plot of total visits to rides in the Kiddie Rides category.

**Note:** For this question, display the box plot in the notebook and print the number of visits to each ride as a list (ex: [3, 4, 5, 6, ...])

````python
# Graded Cell, PartID: zdzaT
# your code here
db.execute("Select attraction, count(*) From checkin Group By attraction")
l = db.fetchall()
freq = pd.DataFrame.from_records(l, columns=['Attraction_ID', 'Freq'])

db.execute("SELECT AttractionID, name FROM attraction as a where a.category LIKE '%Kiddie%' order by AttractionID;")
a_list =  db.fetchall()
attraction_list = pd.DataFrame.from_records(a_list, columns=['Attraction_ID', 'Ride'])

new_df=attraction_list.merge(freq[['Attraction_ID', 'Freq']])
new_df = new_df.drop(['Attraction_ID'], axis =1)

plt.boxplot(new_df['Freq'].tolist())
plt.title("Total Ride Visits in the Kiddie Rides category")
plt.ylabel("Visit count")
plt.xticks([])
plt.show()
print(new_df['Freq'].tolist())
````
